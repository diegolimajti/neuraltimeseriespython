The discrete time Fourier transform can be conceptualized as the dot product between the data and a series of complex sine waves. The number of sine waves used is determined by $n$, then number of data points. The amplitude for a given frequency $X_f$ is determined by:

$$
X_f = \sum_{k=1}^{n} x_k e^{-i 2 \pi f (k-1)n^{-1}}
$$

Where
$X_f$ is the amplitude of frequency $f$

$x_k$ is the original data at time point k

$e^{-i 2 \pi f (k-1)n^{-1}}$ is a complex sine wave of freqency f at time point k

Implementing the Fourier transform

```python
def nts_dft(data):
    """
    --Input--
    data - Time series to be decomposed
    
    --Return--
    fourier - Result of convolving data with complex sine wave. After the function call,
    the result still needs to be divided by n (# data points), and the x-axis of the frequency
    domain needs to be set to 0...n/2 (nyquist frequency) stepped by srate/2. Also, get only
    absolute value of the amplitude.
    """
    
    #number of time points
    n = data.shape[0]
    fourier = np.array([0+0j for i in range(n)])
    
    #normalize time to 0-1 scale
    time = np.arange(start = 0, stop = n, step = 1.) / n
        
    #I'm starting at n=1 to be consistent with the DFT equation
    for i in range(1,n):
        #create sine wave of frequency i-1
        sine_wave = np.e ** ((-1j * 2 * np.pi * (i-1))* time)
        
        #compute the dot product between sine wave and data
        fourier[i-1] = sum(sine_wave * data)
    
    return fourier
```

The inverse Fourier transform is the use of the amplitude data contained in the frequency domain to reconstruct the signal in the time domain. It can be represented by:

$$
x_k = \sum_{k=1}^{n} X_k e^{i 2 \pi f (k-1) n^{-1}}
$$

Where

$x_k$ is the signal in the time domain

$X_k$ is the coefficient by which frequency $f$ is multiplied

$e^{i 2 \pi f (k-1) n^{-1}}$ is the complex sine wave of frequency $f$

Implementing the inverse Fourier transform

```python
def nts_inverse_dft(fourier):
    """
    --Input--
    fourier - data in the Frequency domain
    
    --Return--
    data - Original data
    """
    n = fourier.shape[0]
    data = np.zeros(n)
    
    #normalize time to 0-1 scale
    time = np.arange(start = 0, stop = n, step = 1.0) / n
    
    for i in range(1,n):
        #increment to the data a sine wave of frequency i-1 scaled by fourier[i]
        data += np.real(fourier[i-1] * np.e ** ((1j * 2 * np.pi * (i-1)) * time))
    
    return data
```

Comparing stationary vs. nonstationary signals (Fig. 11.9)

```python
freqs = [3, 10, 5, 7]
amps = [5, 15, 10, 5]
phases = [np.pi / 7, np.pi / 8, np.pi, np.pi/2]

#create time axis
srate = 500.0
time = np.arange(start = -1.0, stop = 1.0, step = 1.0/srate)

#create stationary and nonstationary waves
stationary = np.zeros(time.shape[0] * len(freqs))
nonstationary = np.zeros(time.shape[0] * len(freqs))
n = stationary.shape[0]

for i in range(len(freqs)):
    temp_sine_wave = amps[i] * np.sin((2 * np.pi * freqs[i] * time) + phases[i])
    stationary += np.tile(temp_sine_wave, len(freqs))
    
    temp_sine_wave = temp_sine_wave * (time + 1)
                          
    start_idx = i * time.shape[0]
    stop_idx = i * time.shape[0] + time.shape[0] 
                          
    nonstationary[start_idx : stop_idx] = temp_sine_wave

f_stationary = np.fft.fft(stationary) / n
f_nonstationary = np.fft.fft(nonstationary) / n

x = np.linspace(start = 0.0, stop = srate / 2.0, num = n/2.0)

fig7 = plt.figure(figsize=(12,8))

plt.subplot2grid(shape = (2,2), loc = (0,0), rowspan=1, colspan=1)
plt.title('Stationary signal')
plt.plot(stationary, color=(.3, .3, .3))
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot2grid(shape = (2,2), loc = (0,1), rowspan=1, colspan=1)
plt.title('Nonstationary signal')
plt.plot(nonstationary, color=(.3, .3, .3))
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot2grid(shape = (2,2), loc = (1,0), rowspan=1, colspan=2)
plt.title('Power spectra from Fourier transform')
p_sta, = plt.plot(x, abs(f_stationary[0:n/2]), color=(.7, .7, .7))
p_nonsta, = plt.plot(x, abs(f_nonstationary[0:n/2]), color=(.3, .3, .3))
plt.xlim([0,15])
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.legend([p_sta, p_nonsta], ['Power: stationary', 'Power: nonstationary'])
```

```python
srate = 1000.0
#time = np.arange(start = -0.5, stop = 0.5 - 1/srate, step = 1/srate)
time = np.arange(start = -0.5, stop = 0.5, step = 1/srate)

fg = [15, 5]
#create sine wave (20Hz)
s = np.sin(2 * np.pi * 20 * time)

#create two gaussians (kernels)
g_narrow = np.e ** ((-time ** 2) / (2*(4/(2*np.pi*15)**2)))/15;
g_wide = np.e ** ((-time ** 2) / (2*(4/(2*np.pi*5)**2)))/5;

#convolve sine wave with gaussians
conv_narrow = np.convolve(s, g_narrow)
conv_wide = np.convolve(s, g_wide)

fig8 = plt.figure(figsize=(14,10))

for i in [1,2]:
    plt.subplot(3,2,i)
    plt.title('Sine wave')
    plt.plot(time, s, color = (.3, .3, .3))
    plt.ylim(-2, 2)
    plt.locator_params(nbins=4)

plt.subplot(3,2,3)
plt.title('Narrow gaussian')
plt.plot(time, g_narrow, color = (.3, .3, .3))
plt.ylabel('Amplitude')
plt.locator_params(nbins=4)
plt.ylim([0, 0.08])

plt.subplot(3,2,4)
plt.title('Wide gaussian')
plt.plot(time, g_wide, color = (.3, .3, .3))
plt.locator_params(nbins=4)
plt.ylim([0, 0.25])

plt.subplot(3,2,5)
plt.title('Convolution with narrow gaussian')
plt.plot(time, conv_narrow[conv_narrow.shape[0]*0.25:conv_narrow.shape[0]*0.75:], color = (.3, .3, .3))
plt.locator_params(nbins=4)
plt.xlabel('Time(s)')

plt.subplot(3,2,6)
plt.title('Convolution with wide gaussian')
plt.plot(time, conv_wide[conv_wide.shape[0]*0.25:conv_wide.shape[0]*0.75:], color = (.3, .3, .3))
plt.locator_params(nbins=4)
plt.xlabel('Time(s)')
```

Exercise (3) Test effect of small and large noise on the Fourier transform

```python
srate = 500.0
freqs = [1,4,7,10]
time = np.arange(start = 0, stop = 4, step = 1/srate)
sine_waves = np.zeros([len(freqs),time.size])

for i in range(4):
    sine_waves[i,:] = np.sin(2 * np.pi * freqs[i] * time)
    
sum_wave = sum(sine_waves)
sum_wave_low = sum_wave + 2 * np.random.random_sample(sum_wave.shape)

sum_wave_high = np.copy(sum_wave)
for i in range(10):
    sum_wave_high += 10 * np.random.random_sample(sum_wave.shape)
sum_wave_high /= 10

f_sum = np.fft.fft(sum_wave)
f_low = np.fft.fft(sum_wave_low)
f_high = np.fft.fft(sum_wave_high)
x = np.linspace(start = 0.0, stop = srate / 2.0, num = f_sum.size/2.0)

fig12 = plt.figure(figsize=(16,12))

titles = ['No noise', 'Low noise', 'High noise']
waves = [sum_wave, sum_wave_low, sum_wave_high]
for i in range(3):
    plt.subplot(3,3,i+1)
    plt.title(titles[i])
    plt.plot(time, waves[i], color=(.3, .3, .3))
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    
waves2 = [f_sum, f_low, f_high]
for i in range(3):
    plt.subplot(3,3,4+i)
    plt.bar(left = x, height = abs(waves2[i][0 : waves2[i].size/2]/waves2[i].size), align='center', color=(.3, .3, .3))
    plt.xlim([0,260])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power')
```

A small amount of noise attenuates the power of the original freqencies and adds a little power for every other frequency up to half the sampling rate. A large amount of noise increased the power of the DC component (0Hz) and decreased the power of every other frequency up to halft the sampling rate.

With noise, it's easier to detect the frequency components in the frequency domain, specially for the low noise case. Although small in power, in the large noise case it's still possible to see the components, while it is impossible to detect the frequencies in the time domain.
