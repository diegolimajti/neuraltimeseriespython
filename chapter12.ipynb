```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import load_neural_data

data, time, channels, srate = load_neural_data.load_data()

#primary color for plots
gray1 = (.3, .3, .3)
gray2 = (.7, .7, .7)
```

A Morlet wavelet is a sine wave windowed by a Gaussian

A gaussian is defined by

$$
G(t) = a e^{-(t-m)^2/(2s^2)}
$$

Where $G(t)$ is the value of the gaussian function at time t

$a$ is a amplitude

$m$ is an a-axis offset (mean), for the current purpuses can be defaulted to 0

$s$ is the standard deviation, defined by 

$$
s = \frac{n}{2 \pi f}
$$

where $n$ is the number of wavelet cycles, relevant for the frequency and time resolution tradeoff

and $f$ is the frequency

A simple 4Hz wavelet (Fig 12.1)

```python
f = 4
time2 = np.arange(start = -1, stop = 1, step = 1/srate)
wave = np.cos(2 * np.pi * f * time2)

s = 4/(2* np.pi *f)
gauss = np.e ** (-time2 ** 2 / (2 * s ** 2))

wavelet = gauss * wave
fig1 = plt.figure(figsize=(8,6))
plt.title('A 4Hz wavelet')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim([-0.8, 1.2])
plt.plot(time2, wavelet, color = gray1)
```

Comparing different approaches for temporal weighting (Fig. 12.2)

```python
fig2 = plt.figure(figsize=(8,12))
fig2.subplots_adjust(hspace = 0.5)

plt.subplot(5,1,1)
plt.title('EEG data')
plt.plot(data[channels.index('FCz'), :, 1], color=gray1)

plt.subplot(5,1,2)
sine = np.cos(2 * np.pi * 12 * time2)
plt.title('No temporal weighting (Fourier transform)')
plt.plot(sine, color = gray1)
plt.xlim(0, 500)

plt.subplot(5,1,3)
boxcar = np.zeros(sine.size)
midpoint = time2.size / 2 - 1
boxcar[midpoint - srate/12/5 : midpoint + srate/12/1.25] = 1
plt.title('Strong temporal weighting')
plt.plot(sine * boxcar, color = gray1)
plt.xlim(0, 500)

plt.subplot(5,1,4)
boxcar = np.zeros(sine.size)
boxcar[midpoint - 50 : midpoint + 50] = 1
plt.title('Boxcar temporal weighting')
plt.plot(sine * boxcar, color = gray1)
plt.xlim(0, 500)

plt.subplot(5,1,5)
s=1.5/(2*np.pi*f)
gaussian_win = np.e ** (-time2 ** 2 / (2 * s ** 2))
plt.title('Gaussian temporal weighting')
plt.plot(sine * gaussian_win, color = gray1)
plt.xlim(0, 500)
```

Demonstrating the construction of a Morlet wavelet (Fig. 12.3)

```python
srate = 500
time3 = np.arange(start = -1, stop = 1, step = 1/srate)
f = 10
sine = np.sin(2 * np.pi * f * time3)
s = 6 / (2 * np.pi * f)
gaussian = np.e ** (-time3 ** 2 / (2 * s ** 2))

fig3 = plt.figure(figsize=(8,8))
fig3.subplots_adjust(hspace = 0.4)

plt.subplot(3,1,1)
plt.title('Sine wave')
plt.plot(sine, color = gray1)

plt.subplot(3,1,2)
plt.title('Gaussian window')
plt.plot(gaussian, color = gray1)

plt.subplot(3,1,3)
plt.title('My first wavelet')
plt.plot(sine * gaussian, color = gray1)
```

Creating multiple wavelets (Fig. 12.4)

```python
frequencies = np.linspace(start = 2, stop = 100, num = 80)
wavelet_family = np.zeros([frequencies.size, time3.size])

for i in range(frequencies.size):
    wave = np.e ** (2 * 1j * np.pi * frequencies[i] * time3) 
    s = 6 / (2 * np.pi * frequencies[i])
    gauss = np.e ** (- time3 ** 2 / (2 * s ** 2))
    wavelet_family[i] = wave * gauss

fig4 = plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('A few wavelets')
plt.xlabel('Time')
plt.ylabel('Amplitude')

#Doing a bit different here to generate the same number of waves
for i in range(4):
    plt.plot(np.real(wavelet_family[round(np.random.random() * 79)]), color = gray1)
    
    
#There should be an off-phase imaginary wave here. I wonder why I'm seeing a flat line.
#Not plotting the "Wavelet family portrait"
plt.subplot(1,2,2)
plt.title('Real and imaginary parts of a wavelet')
p_real, = plt.plot(np.real(wavelet_family[30]), color = gray1)
p_imag, = plt.plot(np.imag(wavelet_family[30]), color = gray2)
plt.legend([p_real, p_imag],['Real', 'Imaginary'])
plt.xlim([400,600])
plt.xlabel('Time')
plt.ylabel('Amplitude')
```
