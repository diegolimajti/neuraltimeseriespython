#### Python version of equivalent online Matlab code

Impementation of dot product:

$$ 
\sum_{i=0}^{n} a_ib_i 
$$

Where $a$ and $b$ are $n$-sized vectors

Using as example two random 10-dimensional vectors a and b

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import load_neural_data

a = np.random.random(10)
b = np.random.random(10)
pairwise_result = np.zeros(10)

for i in range(pairwise_result.shape[0]):
    pairwise_result[i] = a[i] * b[i]
    
dotproduct = sum(pairwise_result)
```

The same result can be achieved by using numpy's dot(), that is equivalent to Matlab operation a' \* b

```python
dotproduct = np.dot(a, b)
```

Implementation of the convolution:

$$
(a * b)_k = \sum_{i=0}^{n} a_ib_{k-i}
$$

Where $(a * b)_k$ is the $k$-th resulting point in the convolved signal

$a$ is the original signal

$b$ is the reversed kernel 

First Using numpy's convolve()

```python
#create the data (square wave)
impfun = np.zeros(100)
impfun[45:55] = 1

#create kernel
kernel = np.array([1, .8, .6, .4, .2])

#calculate convolution
conv_numpy = np.convolve(a = impfun, v = kernel, mode = 'valid')

#scale result
conv_numpy_scaled = conv_numpy / sum(kernel)
```

Comparing the original signal, the kernel and the (scaled) result of convolve() (Fig. 10.2)

```python
fig1 = plt.figure(figsize=(10,8))
fig1.subplots_adjust(hspace=0.3)

fig1.add_subplot(3,1,1)
plt.title('Signal')
plt.xlim([0, 100])
plt.ylim([-.1, 1.1])
plt.plot(impfun, color=(.3,.3,.3))

fig1.add_subplot(3,1,2)
plt.title('Kernel')
plt.xlim([0, 100])
plt.ylim([-.1, 1.1])
plt.plot(kernel,  color=(.3,.3,.3))

fig1.add_subplot(3,1,3)
plt.title('Result of convolution')
plt.xlim([0, 100])
plt.ylim([-.1, 1.1])
plt.plot(conv_numpy_scaled,  color=(.3,.3,.3))
```

And now by a manual implementation of a convolution


```python
def nts_convolution(function, kernel):
    """
    --Parameters--
    function - Numpy array to be convolved
    kernel - Numpy array which will work as the kernel for convolution
    
    --Returns--
    conv_result - Numpy array representing the convolved signal
    """
    
    #add <kernel size -1> zeros at start and end of signal
    zeros_padding = np.zeros(kernel.shape[0] -1)
    data_for_conv = np.concatenate((zeros_padding, function, zeros_padding))

    #calculate half kernel size
    half_kernel_size = int(math.ceil((kernel.shape[0]-1) / 2))

    #create array to hold the results (size of original data + kernel size,
    #since the result will have <kernel size / 2> extra points at each end)
    conv_result = np.zeros(function.shape[0] + kernel.shape[0] - 1)

    #calculate a dot product for each time point of this new array
    for i in range(conv_result.shape[0] - half_kernel_size):
        conv_result[i] = np.dot(data_for_conv[i : i+kernel.shape[0]], kernel[::-1])

    #slice this array to get rid of the extra points
    conv_result = conv_result[half_kernel_size + 1 : conv_result.shape[0] - half_kernel_size]

    return conv_result
```

```python
conv_result = nts_convolution(impfun, kernel)
```

Now comparing results of the manual implementation with numpy's convolve(). (Fig. 10.4)

```python
fig2 = plt.figure(figsize=(10,6))
plt.title('Comparing results of manual and numpy convolution')
plt.ylim(-0.1, 3.1)

#plot original signal
signal_plot, = plt.plot(impfun, color = (0, 0, 0))

#plot result of manual convolution 
man_plot, = plt.plot(conv_result, '--', color=(.7,.7,.7))

#plot scaled result of manual convolution 
man_plot_scaled, = plt.plot(conv_result / sum(kernel), color=(.7,.7,.7),)

#plot result of numpy's convolve. It returned me an array of size 96 (maybe it isn't zero-padding?).
#I'm adding 0s at start and end of the array to solve the discrepancies
numpy_plot_scaled, = plt.plot(np.concatenate((np.zeros(1), conv_numpy_scaled, np.zeros(3))),
                              'o', markeredgecolor='None',color = (.7, .7, .7))

plt.legend([signal_plot, man_plot, man_plot_scaled, numpy_plot_scaled],
['Original signal', 'Manual convolution', 'Manual convolution scaled', 'Numpy convolution scaled'])
```

#### Exercises

Creating the kernels

```python
#both kernels are size 11
inv_u_kernel = np.array([-(i * 0.1) ** 2 + 0.25 for i in range (-5,6)])
decay_kernel = np.array([1.0 / (2 * i) for i in range(1,12)])
```

Plotting the kernels

```python
fig3 = plt.figure(figsize=(12,4))
fig3.subplots_adjust(wspace=0.4)

fig3.add_subplot(1,2,1)
plt.title('\"Inverted U\" kernel')
plt.xlim([-0.5, 10.5])
plt.ylim([-0.01, 0.26])
plt.plot(inv_u_kernel,'o', color=(.7,.7,.7), markeredgecolor='None')

fig3.add_subplot(1,2,2)
plt.title('Decay kernel')
plt.xlim([-0.5, 10.5])
plt.ylim([-0.01, 0.26])
plt.plot(decay_kernel,'o', color=(.7,.7,.7), markeredgecolor='None')
```

Loading EEG data and convolving with the kernels

```python
data, time, channels = load_neural_data.load_data()
```

```python
#get data at Fz : use all time points : at trial 1
fz_t1 = data[channels.index('Fz'), :, 0]

fz_t1_inv_u = nts_convolution(function = fz_t1, kernel = inv_u_kernel) / sum(inv_u_kernel)
fz_t1_decay = nts_convolution(function = fz_t1, kernel = decay_kernel) / sum(decay_kernel)
```

Showing the results

```python
fig4 = plt.figure(figsize=(12,8))
fig4.subplots_adjust(hspace=0.6)

fig4.add_subplot(3,1,1)
plt.title('Original signal at Fz in trial 1')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage ((uV)')
plt.ylim([-50, 50])
plt.plot(time, fz_t1, color = (.5, .5, .5))

fig4.add_subplot(3,1,2)
plt.title('Convolved with \"Inverted U\" kernel')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage ((uV)')
plt.ylim([-50, 50])
#My nts_convolve is returning a 639 array while the original data is a 640 array.
#I'm cutting of the last time sample to plot here, but I could be shifting everying by 1 sample
plt.plot(time[:-1], fz_t1_inv_u, color = (.5, .5, .5))

fig4.add_subplot(3,1,3)
plt.title('Convolved with decay kernel')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage ((uV)')
plt.ylim([-50, 50])
#Same problem as in the last one
plt.plot(time[:-1], fz_t1_decay, color = (.5, .5, .5))
```

By visual inspection, it's possible to notice that the "inverted U" kernel is attenuating the signal more than the decay kernel.
